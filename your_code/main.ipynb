#### What's the mean salary by department?
Hint : There are 2 ways to do it - one way is with groupby and the other way is with pivot tables. Do it both ways. 


```python
# Method 1
# your answer here
display(pd.DataFrame(employee.groupby('Department')['Salary'].agg(np.mean).reset_index()))
employee.pivot_table(index= ['Department'], values = ['Salary'])
```

#### We want now to group the data by more than one column. Use Pandas pivot table function to select the mean salary by department and title.

```python
# Method 2
# your answer here
display(pd.DataFrame(employee.groupby(['Department','Title'])['Salary'].agg(np.mean).reset_index()))
employee.pivot_table(index= ['Department','Title'], values = ['Salary'])
```

#### We want to know the information about the mean salary and number of employees in each department and for each job title.

### Bonus

#### We now want to compute the mean salary after removing the maxima for each Department. Create the appropriate pivot table.

Hint: Write a custom function.

```python
# your answer here
employee_max = employee.pivot_table(index= ['Department'], values = ['Salary'], aggfunc=['max']).reset_index()
employee_max.columns = ['_'.join(col) for col in employee_max.columns.values] # jewell of the crown :)
employee_max
```

```python
employee_m = employee.merge(employee_max,how='outer', left_on='Department', right_on='Department_')
employee_m
```

```python
employee_m[employee_m['Salary'] == employee_m['max_Salary']] # Max salary people    
```

```python
employee_m[employee_m['Salary'] != employee_m['max_Salary']] # Not Max salary
```
